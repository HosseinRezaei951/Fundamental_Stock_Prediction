```python
import pandas as pd 
import numpy as np

dataFiles_name = "Main_Data"

df = pd.read_csv(f'{dataFiles_name}.csv')
new_df = df.dropna()

groupBase_column_set = set(new_df["CompanyId"])
for x in groupBase_column_set:
    if len(new_df[new_df["CompanyId"]==x]) < 2:
        new_df = new_df.drop(new_df[new_df["CompanyId"] == x].index)
new_df
```

```python
df_X = new_df[[x for x in new_df.columns if x not in ["CompanyId", "PersianYear", "Return", "Beta"]]]
df_X
```

```python
c ={}
temp = df_X.corr()
for column in temp:
    c[column] = { "n" :  0, 'cols' : [] } 
    for i, val in enumerate(temp[column]):
        if abs(val) > 0.95 and abs(val) != 1:
            # print(abs(val))
            c[column]["n"] += 1
            c[column]["cols"].extend([df_X.columns[i]])

i = 1
for k, v in c.items():
    if v["n"] > 0:
        print(f'{i} {k} : {v["n"]} {v["cols"]}')
        i += 1
```
