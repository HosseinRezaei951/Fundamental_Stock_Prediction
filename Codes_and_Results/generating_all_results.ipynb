```python
import os
import shutil

def get_directories(path):
    # Get all items in the directory
    items = os.listdir(path)

    # Filter out directories
    directories = [item for item in items if os.path.isdir(os.path.join(path, item))]

    return directories


# Get the current working directory (where the script is running)
current_directory = os.getcwd()

AllResults_path = f"{current_directory}/AllResults"
if not os.path.isdir(AllResults_path):
    os.makedirs(AllResults_path)

# Get a list of all directory names
results_directories = get_directories(current_directory)

if "AllResults" in results_directories:
    results_directories.remove("AllResults")
    
print(results_directories)

for dir in results_directories:
    shutil.copytree(f"{current_directory}/{dir}/FinalResults", f"{AllResults_path}/{dir}/", dirs_exist_ok=True)
```

```python
finalResults_path = f"{AllResults_path}/FinalResults"
if not os.path.isdir(finalResults_path):
    os.makedirs(finalResults_path)

# Get a list of all directory names
finalResults_directories = get_directories(AllResults_path)

if "FinalResults" in finalResults_directories:
    finalResults_directories.remove("FinalResults")
    
print(finalResults_directories)
```

```python
preprocessing_methods = ["MinMax", "Standard"]
target_columns = ["r_dicho", "b_dicho"]


bar_plot_path = f"{finalResults_path}/bar_plot"
if not os.path.isdir(bar_plot_path):
    os.makedirs(bar_plot_path)

KBest_rankedFeatures_path = f"{finalResults_path}/KBest_rankedFeatures"
if not os.path.isdir(KBest_rankedFeatures_path):
    os.makedirs(KBest_rankedFeatures_path)

for dir in finalResults_directories:
    tmp_filePath = f"{AllResults_path}/{dir}"
    tmp_fileName = f"bar_plot.png"
    shutil.copy(f"{tmp_filePath}/{tmp_fileName}", f"{bar_plot_path}/{dir}.png")
    
    tmp_filePath = f"{AllResults_path}/{dir}"
    tmp_fileName = f"final_KBest_rankedFeatures.csv"
    shutil.copy(f"{tmp_filePath}/{tmp_fileName}", f"{KBest_rankedFeatures_path}/{dir}.csv")
    
    for target in target_columns:        
        
        max_path = f"{finalResults_path}/max/{target}"
        if not os.path.isdir(max_path):
            os.makedirs(max_path)

        mean_path = f"{finalResults_path}/mean/{target}"
        if not os.path.isdir(mean_path):
            os.makedirs(mean_path)

        auc_roc_mean_path = f"{finalResults_path}/auc_roc_mean/{target}"
        if not os.path.isdir(auc_roc_mean_path):
            os.makedirs(auc_roc_mean_path)

        aupr_mean_path = f"{finalResults_path}/aupr_mean/{target}"
        if not os.path.isdir(aupr_mean_path):
            os.makedirs(aupr_mean_path)

        for pp_method in preprocessing_methods:           

            tmp_filePath = f"{AllResults_path}/{dir}"
            tmp_fileName = f"{target}_max.csv"
            shutil.copy(f"{tmp_filePath}/{tmp_fileName}", f"{max_path}/{dir}.csv")

            tmp_filePath = f"{AllResults_path}/{dir}"
            tmp_fileName = f"{target}_mean.csv"
            shutil.copy(f"{tmp_filePath}/{tmp_fileName}", f"{mean_path}/{dir}.csv")

            tmp_filePath = f"{AllResults_path}/{dir}"
            tmp_fileName = f"{target}_{pp_method}_auc_roc_mean.png"
            shutil.copy(f"{tmp_filePath}/{tmp_fileName}", f"{auc_roc_mean_path}/{dir}_{pp_method}.png")

            tmp_filePath = f"{AllResults_path}/{dir}"
            tmp_fileName = f"{target}_{pp_method}_aupr_mean.png"
            shutil.copy(f"{tmp_filePath}/{tmp_fileName}", f"{aupr_mean_path}/{dir}_{pp_method}.png")
```

```python
import pandas as pd

final_df = pd.DataFrame()

for dir in finalResults_directories:
    for target in target_columns:  
        tmp_filePath = f"{AllResults_path}/{dir}"
        tmp_fileName = f"{target}_mean.csv" 
        tmp_df = pd.read_csv(f"{tmp_filePath}/{tmp_fileName}")
        tmp_df.insert(0, "Target", f"{target}")
        tmp_df.insert(0, "Run_sample", f"{dir}")
        final_df = pd.concat((final_df, tmp_df))

final_df = final_df.reset_index(drop=True)
final_df
# final_df.style.background_gradient(cmap="Reds", axis=False)
```

```python
# Function to extract mean and std components and convert to numeric types
def extract_and_convert(value):
    mean, std = map(float, value.split(' ± '))
    return mean, std

metrics = [
    'Acc Cv', 'Acc Te', 'Sp', 'Se',
    'MCC', 'F1', 'AUC-ROC', 'AUPR'
]

new_final_df = pd.DataFrame()
for i, col in enumerate(final_df.columns):
    if col not in metrics:
        new_final_df[col] = final_df[col].values.copy()

    else:
        new_final_df[[f"{col}_mean", f"{col}_std"]] = final_df[col].apply(extract_and_convert).apply(pd.Series)

new_final_df
```

```python
new_final_df = new_final_df.sort_values(by="AUPR_mean", ascending=False, ignore_index=True)
new_final_df.to_csv("final_df.csv", index=False)
new_final_df
```

```python
for target in target_columns:
    tmp_df = new_final_df.groupby("Target").get_group(target).reset_index(drop=True)
    tmp_df.drop(columns="Target", inplace=True)
    tmp_df.to_csv(f"{target}_df.csv", index=False)
```

```python
r_dicho_df = pd.read_csv("r_dicho_df.csv")

alg_divided = r_dicho_df.sort_values(by=['AUPR_mean', 'Scaling', 'Alg'], ascending=False).reset_index(drop=True)
alg_divided = alg_divided[alg_divided["Run_sample"].isin(["new_stat", "new_conv"])].reset_index(drop=True)
alg_divided = alg_divided[alg_divided["Scaling"].isin(["Standard", "MinMax"])].reset_index(drop=True)
alg_divided = alg_divided.drop(columns=[
    'Acc Cv_mean', 'Acc Cv_std',
    'Acc Te_mean', 'Acc Te_std'
])

for col in alg_divided.columns:
    if "_std" in col:
        alg_divided = alg_divided.drop(columns=col)

alg_divided.columns = [
    'Method', 'Scaling', 'Alg', 
    'Sp', 'Se', 'MCC',
    'F1', 'AUC-ROC', 'AUPR'
]
alg_divided = alg_divided[[
    'Method', 'Alg', 'Scaling', 
    'Sp', 'Se', 'MCC',
    'F1', 'AUC-ROC', 'AUPR'
]]
for i in range(len(alg_divided)):
    if alg_divided.loc[i, "Method"] == "new_stat":
        alg_divided.loc[i, "Method"] = "Stat"
    else:
        alg_divided.loc[i, "Method"] = "Conv"

alg_divided = alg_divided[alg_divided["Alg"] != "DT"]

alg_divided.set_index(["Method", "Alg", "Scaling"], inplace=True)

s = alg_divided.style\
    .background_gradient(cmap="Oranges", subset=alg_divided.columns[0:]) \
    .format('{:.2f}', subset=alg_divided.columns[0:])

s.to_latex(
    "r_dicho_table.txt",
    caption="Selected stock correlation and simple statistics.",
    clines="skip-last;data",
    convert_css=True,
    position_float="centering",
    multicol_align="|c|",
    hrules=True,
)  
s
```

```python
%matplotlib inline
import re
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
import pandas as pd
import matplotlib as mpl
import numpy as np

# Set the custom font settings for scientific paper publications
mpl.rcParams.update({
    'font.family': 'serif',
    'font.serif': ['Times New Roman'],
    'font.size': 18,
    'font.weight': 'normal',
    'axes.labelsize': 18,
    'axes.titlesize': 18,
    'axes.titleweight': 'bold',
    'xtick.labelsize': 18,
    'ytick.labelsize': 18,
    'legend.fontsize': 16,
})

stat_path = "new_stat/Results/3_Algorithms_Results/r_dicho"
# Get a list of all directory names
stat_directories = get_directories(stat_path)
stat_directories

stat_df = pd.DataFrame()

for dir in stat_directories:
    if "_Standard" in dir or "_MinMax" in dir:
        tmp_filePath = f"{stat_path}/{dir}"
        tmp_file = pd.read_csv(f"{tmp_filePath}/results_df.csv")[["Algorithm", "Number_of_Features", "aupr_mean"]]
        tmp_file.insert(0, "Scaling", re.sub(r'\d+_', '', dir)) 
        stat_df = pd.concat([stat_df, tmp_file])

stat_df = stat_df[stat_df["Algorithm"] != "DT"]
stat_df = stat_df.groupby(by=["Algorithm", "Scaling", "Number_of_Features"]).mean(numeric_only=True)

stat_df.to_csv("stat_df.csv")


metric = "aupr_mean"
# Define the line styles for each line
line_styles = ["-", "--", "-.", ":"]
marker_styles = ["o", "^", "D", "s"]
number_of_features = 188

fig = plt.figure(figsize=(16, 9))
gs = GridSpec(1, 1)

ax_scores = fig.add_subplot(gs[0, 0])

# Plot each series with the corresponding line style
for i, (alg, scaling) in enumerate(stat_df[metric].unstack(level=[0, 1]).columns):
    ax_scores.plot(
        stat_df[metric].loc(axis=0)[(alg, scaling)].index[:number_of_features],
        stat_df[metric].loc(axis=0)[(alg, scaling)].values[:number_of_features],
        line_styles[i % len(line_styles)],  # Cycle through the line styles
        marker=marker_styles[i % len(marker_styles)],  # Cycle through the marker styles
        label=f"{alg} - {scaling}"
    )
    
# show grid lines
tick_positions = np.arange(0, number_of_features + 1, 10)
tick_positions = np.insert(tick_positions, 1, 1)
if tick_positions[-1] != number_of_features:
    tick_positions = np.append(tick_positions, number_of_features)

ax_scores.set_xticks([i for i in range(1, number_of_features+1)])
ax_scores.set_xticklabels(['' if i not in tick_positions else str(i) for i in range(1, number_of_features+1)])

title = metric.replace("_", " ").title()
if "Aupr" in title:
    title = title.replace("Aupr","AUPR")

# ax_scores.set_title(title)
ax_scores.set_xlabel('Number of Features')
ax_scores.set_ylabel('Score (%)')
ax_scores.grid(True)

# show the legend
ax_scores.legend(loc='lower right')

# show/save the plot
plt.tight_layout()
plt.savefig(
    f"r_dicho_stat.png",
    bbox_inches='tight', facecolor='white', dpi=120
)
plt.close()
```
